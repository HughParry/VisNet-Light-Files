```python
from PIL import Image, ImageDraw
import numpy as np

#This file is used to generate crosses of a given thickness for use in the project. Background is set to 127 for use in VisNet

def drawlines(x, y, savename, width):
    
    #Defining initial coordinates. Cross_1 is on the diagonal and Cross_2 is on the vertical/horizontal
    linecoords = {'Cross_1':[(14+x, 14+y, 72+x, 72+y), (72+x, 14+y, 14+x, 72+y)],
                 'Cross_2':[(4+x, 256//6+y, 256//3+x, 256//6+y), (256//6+x, (256)//3 - 2+y, 256//6+x, 4+y)]}
    
    #Plotting Lines
    for coords in linecoords:
        im = Image.new('L', (256, 256), color=127)
        draw = ImageDraw.Draw(im)
        
        draw.line(linecoords[coords][0], fill=(255), width=width)
        draw.line(linecoords[coords][1], fill=(255), width=width)
        
        #Comment out whichever of the following you don't want. im.save prints in .png file format, saveuint8 saves in unsigned ints
        #im.save(savename + coords + '.png')
        saveuint8(im, savename + coords)
        
        
#Func for saving a PIL image as a UINT8 file (i.e. unsigned bytes in the range 0-255)
def saveuint8(image, name):
    
    #Convest PIL Image to numpy array
    numpyimage = np.array(image)

    #Saving numpy array as unsigned bits
    newFile = open(name, "wb")
    newFileByteArray = bytearray(numpyimage)
    newFile.write(newFileByteArray)
    
    
def transformed_lines(width):
    
    #Tidy up spaghetti code function
    
    #Top Lines
    drawlines(0, 0, str(width) + 'Top_Left_', width)
    drawlines(0, 256//3, str(width) + 'Middle_Left_', width)
    drawlines(0, 2*256//3, str(width) + 'Bottom_Left_', width)

    #Middle Lines
    drawlines(256//3, 0, str(width) + 'Top_Center_', width)
    drawlines(256//3, 256//3, str(width) + 'Middle_Center_', width)
    drawlines(256//3, 2*256//3, str(width) + 'Bottom_Center_', width)

    #Bottom Lines
    drawlines(2*256//3, 0, str(width) + 'Top_Right_', width)
    drawlines(2*256//3, 256//3, str(width) + 'Middle_Right_', width)
    drawlines(2*256//3, 2*256//3, str(width) + 'Bottom_Right_', width)
    

#Drawing lines of varying thickness
for width in [2*i for i in range(1, 5)]:  
    transformed_lines(width)
```
