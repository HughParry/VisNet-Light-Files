```python
from PIL import Image, ImageDraw
import numpy as np

#This file is for generating T-Junction primatives in 9 locations i

def drawt(x1, y1, x2, y2, x3, y3, x4, y4, strname, width):

    im = Image.new('L', (256, 256), color=127)
    draw = ImageDraw.Draw(im)
    
    draw.line((x1, y1, x2, y2), fill=255, width=width)
    draw.line((x3, y3, x4, y4), fill=255, width=width)
    
    #im.save(str(width) + strname, quality=95) #Comment out the one of these that you don't want
    saveuint8(im, str(width) + strname[:-3])

    
def saveuint8(image, name):
    
    #Convest PIL Image to numpy array
    numpyimage = np.array(image)

    newFile = open(name, "wb")
    newFileByteArray = bytearray(numpyimage)
    newFile.write(newFileByteArray)
    

def draw_transformed_lines(offsets, width, names):
    
    counter = 0
    
    for x in offsets:
        for y in offsets:
            
            drawt(256//6+x, 256//6+y, 72+x, 72+y, 72+x, 14+y, 14+x, 72+y, names[counter] + 'T_1.jpg', width)
            drawt(14+x, 14+y, 256//6+x, 256//6+y, 72+x, 14+y, 14+x, 72+y, names[counter] + 'T_2.jpg', width)
            drawt(14+x, 14+y, 72+x, 72+y, 256//6+x, 256//6+y, 14+x, 72+y, names[counter] + 'T_3.jpg', width)
            drawt(14+x, 14+y, 72+x, 72+y, 72+x, 14+y, 256//6+x, 256//6+y, names[counter] + 'T_4.jpg', width)
            drawt(4+x, 256//6+y, (256)//3 + 2+x, 256//6+y, 256//6+x, (256)//3 + 2+y, 256//6+x, 256//6+y, names[counter] + 'T_5.jpg', width)
            drawt(2+x, 256//6+y, (256)//3 + 2+x, 256//6+y, 256//6+x, (256)//6+y, 256//6+x, 4+y, names[counter] + 'T_6.jpg', width)
            drawt((256)//6+x, 256//6+y, (256)//3 + 2+x, 256//6+y, 256//6+x, (256)//3 + 2+y, 256//6+x, 4+y, names[counter] + 'T_7.jpg', width)
            drawt(2+x, 256//6+y, (256)//6+x, 256//6+y, 256//6+x, (256)//3 + 2+y, 256//6+x, 4+y, names[counter] + 'T_8.jpg', width)
            
            counter += 1
    
offsets = [0, 256//3, 2*256//3]
names = ['Top_Left_', 'Middle_Left_', 'Bottom_Left_', 'Top_Center_', 'Middle_Center_',
            'Bottom_Center_', 'Top_Right_', 'Middle_Right_', 'Bottom_Right_']


for width in [2*i for i in range(1, 5)]:
    draw_transformed_lines(offsets, width, names)
```
