```python
from PIL import Image
from PIL import ImageDraw
import numpy as np


#Function to return the 'nth' row of Pascal's Triangle
def pascal_row(n, memo={}):
    
    if n in memo:
        return memo[n]
    
    result = [1]
    
    x, numerator = 1, n
    
    for denominator in range(1, n//2+1):
        
        x *= numerator
        x /= denominator
        
        result.append(x)
        numerator -= 1
        
    if n&1 == 0:
        result.extend(reversed(result[:-1]))
        
    else:
        result.extend(reversed(result))
        
    memo[n] = result
    return(result)


#TODO - Tidy up this code to make it less awful
def draw_bezier(coords): #coords tuples that form the command points
    
    n = len(coords)
    combinations = pascal_row(n-1)
    
    def calc_bezier(ts):
        
        result = []
        for t in ts:
            
            tpowers = (t**i for i in range(n))
            upowers = reversed([(1-t)**i for i in range(n)])
            
            coefs = [c*a*b for c, a, b in zip(combinations, tpowers, upowers)]
            result.append(tuple(sum([coef*p for coef, p in zip(coefs, ps)]) for ps in zip(*coords)))
                          
        return(result)
    return(calc_bezier)


#TODO - turn these common functions into a mini library to speed up project + avoid repitition etc.
def saveuint8(image, name):
    
    #Turn image into numpy array
    numpyimage = np.array(image)
    
    #Save numpy array as unsigned ints
    newFile = open(name, "wb")
    newFileByteArray = bytearray(numpyimage)
    newFile.write(newFileByteArray)
```

```python
#Now generating the various types of and rotations of Bezier curves for use in the project.

for i in range(1):                 #Y Axis
    for j in range(11):             #X Axis
        for k in range(9):          #Rotation
            
            ts = [t/100.0 for t in range(101)]
            im = Image.new('L', (256, 256), color=127) 
            draw = ImageDraw.Draw(im)

            coords = [(256//3 + 10, 256/2), (256/2 + 8*j, 256/2 + 8*i), (2*256//3 - 10, 256/2)]    #Edit these as required for curve generation
            
            bezier = draw_bezier(coords)
            points = bezier(ts)

            draw.line(points, fill = 256, width=5)
            im = im.rotate(40*k, fillcolor=127)
            
            #Comment out the save commands as required
            im.save('X' + str(i)+'_Y'+str(j)+'_R'+str(40*k)+'.png')
            #saveuint8(im, 'X' + str(i)+'_Y'+str(j)+'_R'+str(40*k))
```
