```python
from PIL import Image, ImageDraw
import numpy as np

#This file generates 'J' or 'L' junctions for use in VisNet. Backgrounds of 127

def drawt(x1, y1, x2, y2, x3, y3, x4, y4, strname, width):
    
    # Top Center Cross 1
    im = Image.new('L', (256, 256), color=127)
    draw = ImageDraw.Draw(im)
    draw.line((x1, y1, x2, y2), fill=255, width=width)
    draw.line((x3, y3, x4, y4), fill=255, width=width)
    
    #Comment out as required
    #im.save(str(width) + strname, quality=95)
    saveuint8(im, str(width) + strname[:-3])


#Save image as unsigned ints
def saveuint8(image, name):
    
    #Convest PIL Image to numpy array
    numpyimage = np.array(image)

    newFile = open(name, "wb")
    newFileByteArray = bytearray(numpyimage)
    newFile.write(newFileByteArray)
    
    
def junctions_transforms(offsets, width, names):
    
    startnames = [str(i) + '.png' for i in range(1,9)]
    counter = 0
    
    for x in offsets:
        for y in offsets:
            
            #Diags
            drawt(256//6 - 3 + x, 256//6 + y, 256//3 + x, 256//6 + y, 256//6 + x, 256//6 + y, 256//6 + x, 4 + y, names[counter] + startnames[0], width)
            drawt(4 + x, 256//6 + y, 256//6 + 2 + x, 256//6 + y, 256//6 + x, (256)//3 - 2 + y, 256//6 + x, 256//6 + y, names[counter] + startnames[1], width)
            drawt(4 + x, 256//6 + y, 256//6 + 2 + x, 256//6 + y, 256//6 + x, 256//6 + y, 256//6 + x, 4 + y, names[counter] + startnames[2], width)
            drawt(256//6 - 3 + x, 256//6 + y, 256//3 + x, 256//6 + y, 256//6 + x, (256)//3 - 2 + y, 256//6 + x, 256//6 + y, names[counter] + startnames[3], width)

            #Not Diags
            drawt(256//6 + x, 256//6 + y, 72 + x, 72 + y, 256//6 + x, 256//6 + y, 14 + x, 72 + y, names[counter] + startnames[4], width)
            drawt(14 + x, 14 + y, 256//6 + x, 256//6 + y, 256//6 + x, 256//6 + y, 14 + x, 72 + y, names[counter] + startnames[5], width)
            drawt(14 + x, 14 + y, 256//6 + x, 256//6 + y, 72 + x, 14 + y, 256//6 + x, 256//6 + y, names[counter] + startnames[6], width)
            drawt(256//6 + x, 256//6 + y, 72 + x, 72 + y, 72 + x, 14 + y, 256//6 + x, 256//6 + y, names[counter] + startnames[7], width)
            
            counter += 1

            
#Names for identifying location within picture for use in VisNet analysis etc
names = ['Top_Left_', 'Middle_Left_', 'Bottom_Left_', 'Top_Center_', 'Middle_Center_',
         'Bottom_Center_', 'Top_Right_', 'Middle_Right_', 'Bottom_Right_']

#Offsets for generating the 9 different 'views' per picture
offsets = [0, 256//3, 2*256//3]

for width in [2*i for i in range(1, 5)]:
    junctions_transforms(offsets, width, names)
```
