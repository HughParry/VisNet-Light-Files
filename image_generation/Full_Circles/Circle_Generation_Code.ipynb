```python
from PIL import Image, ImageDraw
import numpy as np

#This script generates circle segments of a given thickness for use in the project

def drawcirc(shape, start, end, strname, width):
    
    im = Image.new("L", (256, 256), color=127)
    draw = ImageDraw.Draw(im)  
    draw.arc(shape, start, end, fill = (256), width=width)
    #im.save(strname, quality=95)
    saveuint8(im, strname[:-3])
    
    
def allcircs(shape, append_str, width):
    drawcirc(shape, 0, 360, str(width) + '_' + append_str + ".jpg", width)

    
def saveuint8(image, name):
    
    #Convest PIL Image to numpy array
    numpyimage = np.array(image)

    newFile = open(name, "wb")
    newFileByteArray = bytearray(numpyimage)
    newFile.write(newFileByteArray)
    

def allcircles(width):
    
    offsets = [0, 256//3, 2*256//3]
    shape = [[(10+i, 10+j), (256//3-10+i, 256//3-10+j)] for i in offsets for j in offsets]


    names = ['Top_Left_', 'Middle_Left_', 'Bottom_Left_', 'Top_Center_', 'Middle_Center_',
            'Bottom_Center_', 'Top_Right_', 'Middle_Right_', 'Bottom_Right_']

    for i in range(9):
        allcircs(shape[i], names[i], width)
        
for width in [2*i for i in range(1,5)]:
    allcircles(width)
```
